```python
#linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
LR=LinearRegression()
mse=cross_val_score(LR,X,Y, scoring='neg_mean_squared_error',cv=5)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Define Ridge regression model
R = Ridge()

# Parameter grid for alpha values
params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50 ,55, 100]}

# GridSearchCV for Ridge regression
ridge_regression = GridSearchCV(R, params, scoring='neg_mean_squared_error', cv=5)

# Fit the model
ridge_regression.fit(X, Y)

# Output the best parameters and score
print("Best Alpha:", ridge_regression.best_params_)
print("Best Score:", ridge_regression.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Define lasso regression model
l = Lasso()

# Parameter grid for alpha values
params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50 ,55, 100]}

# GridSearchCV for lasso regression
lasso_regression = GridSearchCV(l, params, scoring='neg_mean_squared_error', cv=5)

# Fit the model
lasso_regression.fit(X, Y)

# Output the best parameters and score
print("Best Alpha:", lasso_regression.best_params_)
print("Best Score:", lasso_regression.best_score_)
```
